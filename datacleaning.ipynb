```python
df['Aplle_ID'] = df['Aplle_ID'].fillna(df['Aplle_ID'].mean())
```

```python
df_new = df.dropna()
```

```python
df_new['Quality'].unique()
```

```python
from os import replace
df_new.Quality = df_new.Quality.str.replace('good','Vgood')
df_new.Quality = df_new.Quality.str.replace('bad','Vbad')
```

```python
df_new['Quality'].unique()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_new
            , orient='v', palette='Set2')
plt.xlabel('Columns')
plt.ylabel('Values')
plt.title('Boxplot for Each Column')

plt.show()
```

```python
numeric_columns = df_new.select_dtypes(include='number').columns

fig, axes = plt.subplots(nrows=len(numeric_columns), ncols=1, figsize=(8, 111))

for i, column in enumerate(numeric_columns):
    sns.boxplot(y=df_new[column], ax=axes[i])
    axes[i].set_title(column)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
